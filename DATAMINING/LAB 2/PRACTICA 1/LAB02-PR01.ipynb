#### Librerias

```python
from Apriori import APriori
import warnings
import pandas as pd
import numpy as np
warnings.filterwarnings('ignore')
```

<font color="blue">Antes de ejecutar el algoritmo apriori con el dataset de películas, probaremos la librería con 2 datasets de prueba, uno estableciendo una frecuencia de soporte mínima y otro con un soporte mínimo, para comprobar el correcto funcionamiento de la librería.</font>

## Dataset dummy 1
<br>
<font color="blue">Establecemos una frecuencia soporte mínima de 2 y una confianza del 10% para poder ver muchas reglas.</font>

```python
algoritmo = APriori()
algoritmo.ExcelPrueba(Fichero="Datos/dummy_1.xlsx")
algoritmo.CalcularReglasDeConfianza(MinimoFreqSop=2, Confianza=10, Echo=True)
```

<font color="blue">Observamos 3 reglas con una confianza del 100%, es decir, que se cumplen siempre:
<ul>
    <li>I1, I5 → I2</li>
    <li>I2, I5 → I1</li>
    <li>I5 → I2, I2</li>
</ul>
    Son unas reglas muy buenas.
</font>

## Dataset películas
<br>
<font color="blue">Primero tenemos que preparar el dataset para que sea admitido por la librería de Apriori.</font>

```python
Dataframe = pd.read_csv("Datos/movies_dataset.csv")
df_generos = pd.DataFrame()
df_generos["generos"] = Dataframe['genres']
df_generos["generos"] = df_generos['generos'].apply(lambda x: np.array(x.split('|')))
print("Hay", len(df_generos), "tuplas.")
```

```python
algoritmo_2 = APriori()
algoritmo_2.Carga(Datos=df_generos, Columna="generos")
algoritmo_2.CalcularReglasDeConfianza(MinimoFreqSop=60, Confianza=80.0, Echo=True)
```

<font color="blue">Tras varias ejecuciones del algoritmo con distintos valores mínimos de frecuencia soporte, escogemos 60. Esto se debe a que si es más bajo, las reglas obtenidas son muy largas y el valor de sus confianzas tienen mucho parecido al resultado final, haciendo que no merezca la pena. En cambio, si es más alto, obtenemos muy pocas reglas que son muy cortas y dan poca información. La confianza se establece en un 80% para obtener solo reglas de mucha calidad. 
<br><br>
Observando las reglas, nos damos cuenta de que tienen relación con la realidad (es decir, son válidos). Por ejemplo: Adventure, Animation, Comedy → Family tiene sentido, al igual que Action, Crime, Drama → Thriller. 
<br><br>
Hemos encontrado buenas reglas.
