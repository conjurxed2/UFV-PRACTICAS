#### Librerias

```python
import numpy as np
import graphviz
import pydot
import pandas as pd
import matplotlib.pyplot as plt
import graphviz
import matplotlib.pyplot as plt
import pydotplus
from sklearn import preprocessing
from sklearn.tree import export_graphviz
from IPython.display import SVG
from graphviz import Source
from IPython.display import display
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

#### Dirección del dataset

```python
DIR_DATASET = "Datos/mushrooms.csv"
```

## 1. Preparación del dataset

<font color="blue">Cargamos el dataset de setas</font>

```python
Datos = pd.read_csv(DIR_DATASET)
Datos.head()
```

### Limpiamos los campos que son nulos.

<font color="blue">Contamos los campos nulos por cada columna.</font>

```python
nulos = pd.DataFrame(Datos.isna().sum(), columns=["Número de nulos"])
nulos
```

<font color="blue">No hay nulos.</font>

<font color="blue">Extraemos del dataset la columna que indica a qué clase pertenece la seta (venenosa o normal) y transformamos los valores categóricos a numéricos para poder utilizar la librería de árboles de decisión.</font>

```python
# Creamos un nuevo dataframe para los datos numéricos.
DatoNumericos = pd.DataFrame()

# Iteramos sobre las columnas del dataframe con los datos categóricos.
for columna in Datos.columns:
    # Introducimos la columna con los datos pasados a numéricos. Esto es necesario pues la librería de sklearn solo
    # trabaja con valores numéricos.
    DatoNumericos[columna] = pd.factorize(Datos[columna])[0]

# Creamos el array con la clase de cada registro.
prediccion = DatoNumericos["Venenosa"].values

# Eliminamos la columna "Venenosa" que no la necesitamos más en el dataframe.
del DatoNumericos["Venenosa"]

#Extraemos del dataset el nombre de cada característica.
columnas = DatoNumericos.columns

# Mostramos el nuevo dataframe.
DatoNumericos.head()
```

<font color="blue">Dividimos el conjunto total de datos en uno para entrenar y otro para validar el modelo.</font>

```python
x_train, x_test, y_train, y_test = train_test_split(DatoNumericos, prediccion, stratify=prediccion, test_size=0.2)
```

## 2. Creación del modelo

### 2.1 Entropía

<font color="blue">Probamos a construir un árbol de decisión con todas las características proporcionadas en el dataset. Emplearemos la entropía (el nivel de desorden/homogéneidad) de los conjuntos clasificados.</font>

```python
# Creamos el árbol.
arbol = DecisionTreeClassifier(criterion='entropy')

# Entrenamos el árbol.
arbol.fit(x_train, y_train)

print("El acierto del conjunto de entrenamiento es del " 
      + "\033[1m" + str(round(arbol.score(x_train, y_train) * 100, 1)) + "%" + "\033[0m".format())

print("El acierto del conjunto de validación es del " 
      + "\033[1m" + str(round(arbol.score(x_test, y_test) * 100, 1)) + "%" + "\033[0m" )
```

<font color="blue">Dibujamos el árbol resultante.</font>

```python
fichero = StringIO()
export_graphviz(arbol, out_file=fichero, class_names=["venenosa", "normal"], feature_names=columnas, filled=True)
grafica = pydotplus.graph_from_dot_data(fichero.getvalue())  
Image(grafica.create_png())
```

<font color="blue">Creamos un diagrama de barras con las variables más influyentes en la clasificación.</font>

```python
num_caracteristicas = columnas.shape[0]
plt.subplots(figsize=(20, 10))
plt.barh(range(num_caracteristicas), arbol.feature_importances_, align='center')
plt.yticks(np.arange(num_caracteristicas), columnas)
plt.xlabel('Importancia de la característica')
plt.ylabel('Característica')
plt.show()
```

<font color="blue">Vemos que las características más importantes son:
<ol>
    <li>Olor.</li>
    <li>Color de las esporas.</li>
    <li>Tamaño de las láminas.</li>
    <li>Tallo raíz.</li>
    <li>Habitat.</li>
    <li>Población.</li>
</ol>
El resto no importan. Podemos eliminarlas
</font>

### 2.2 GINI

```python
# Creamos el árbol.
arbol = DecisionTreeClassifier(criterion='gini')

# Entrenamos el árbol.
arbol.fit(x_train, y_train)

print("El acierto del conjunto de entrenamiento es del " 
      + "\033[1m" + str(round(arbol.score(x_train, y_train) * 100, 1)) + "%" + "\033[0m".format())

print("El acierto del conjunto de validación es del " 
      + "\033[1m" + str(round(arbol.score(x_test, y_test) * 100, 1)) + "%" + "\033[0m" )
```

<font color="blue">Dibujamos el árbol resultante.</font>

```python
fichero = StringIO()
export_graphviz(arbol, out_file=fichero, class_names=["venenosa", "normal"], feature_names=columnas, filled=True)
grafica = pydotplus.graph_from_dot_data(fichero.getvalue())  
Image(grafica.create_png())
```

<font color="blue">Creamos un diagrama de barras con las variables más influyentes en la clasificación.</font>

```python
num_caracteristicas = columnas.shape[0]
plt.subplots(figsize=(20, 10))
plt.barh(range(num_caracteristicas), arbol.feature_importances_, align='center')
plt.yticks(np.arange(num_caracteristicas), columnas)
plt.xlabel('Importancia de la característica')
plt.ylabel('Característica')
plt.show()
```

<font color="blue">Vemos que las características más importantes son:
<ol>
    <li>Olor.</li>
    <li>Tallo raíz.</li>
    <li>Tamaño láminas.</li>
</ol>
El resto importan extremadamente poco o nada. Podemos eliminarlas.
</font>

##  3. Modelo final

<font color="blue">Para la construcción del modelo final emplearemos como criterio de división ENTROPY debido a que marca claramente 2 variables como las más importantes para clasificar (Olor y color de las esporas). Para crear el árbol más sencillo, evaluaremos la precisión del árbol solo con las 2 variables más importantes: el olor y el color de las esporas.
<br>
Preparamos el dataset final solo con esas variables.
</font>

```python
Dataset = pd.DataFrame()

Dataset["Olor"] = Datos["Olor"]
Dataset["Esporas_color"] = Datos["Esporas_color"]
Dataset["Venenosa"] = Datos["Venenosa"]    

# Transformación a datos numéricos de la columna "Olor".
encoderOlor = preprocessing.LabelEncoder()
encoderOlor.fit(Dataset["Olor"])
Dataset["Olor"] = encoderOlor.transform(Dataset["Olor"])

# Transformación a datos numéricos de la columna "Venenosa".
encoderVenenoso = preprocessing.LabelEncoder()
encoderVenenoso.fit(Dataset["Venenosa"])
Dataset["Venenosa"] = encoderVenenoso.transform(Dataset["Venenosa"])

# Transformación a datos numéricos de la columna "Esporas_color".
encoderEsporasColor = preprocessing.LabelEncoder()
encoderEsporasColor.fit(Dataset["Esporas_color"])
Dataset["Esporas_color"] = encoderEsporasColor.transform(Dataset["Esporas_color"])

# Creamos el array con la clase de cada registro.
prediccion = Dataset["Venenosa"].values

# Eliminamos la columna "Venenosa" que no la necesitamos más en el dataframe.
del Dataset["Venenosa"]

#Extraemos del dataset el nombre de cada característica.
columnas = Dataset.columns

x_train, x_test, y_train, y_test = train_test_split(Dataset, prediccion, stratify=prediccion, test_size=0.2)

Dataset["Venenosa"] = prediccion

Dataset.head()
```

<font color="blue">Creamos el árbol.</font>

```python
# Creamos el árbol.
arbol = DecisionTreeClassifier(criterion='entropy', max_depth=4)

# Entrenamos el árbol.
arbol.fit(x_train, y_train)

print("El acierto del conjunto de entrenamiento es del " 
      + "\033[1m" + str(round(arbol.score(x_train, y_train) * 100, 1)) + "%" + "\033[0m".format())

print("El acierto del conjunto de validación es del " 
      + "\033[1m" + str(round(arbol.score(x_test, y_test) * 100, 1)) + "%" + "\033[0m" )

fichero = StringIO()
export_graphviz(arbol, out_file=fichero, class_names=["Normal", "Venenosa"], feature_names=columnas, filled=True)
grafica = pydotplus.graph_from_dot_data(fichero.getvalue())  
Image(grafica.create_png())
```

<font color="blue">El árbol usando como criterio de división ENTROPY y con solo las 2 variables más importantes para la clasificación (olor y color de las esporas), conseguimos una precisión del 98% para el conjunto de validación, que es un resultado muy bueno.
</font>

#### Reglas extraídas.
##### Previamente, debido a que la librería de sklearn solo trabaja con valores numéricos, transformamos los valores del dataset a números. Ahora, tenemos que invertir la transformación de variables numéricas a su valor categórico para poder expresar las reglas de forma entendible.
