### Objetivo 

El problema del terrorismo es actualmente considerado uno de los más graves para la población. Es un hecho que este ha ido cambiando a lo largo de la historia. Antaño los grupos terroristas actuaban en su propio país o en los países con los que compartía frontera. Hoy en día el terrorismo se considera como un problema global en el que prácticamente cualquier país puede sufrir sus consecuencias. En esta práctica intentaremos entender varias situaciones tanto actuales cómo históricas. Para ello haremos uso de un dataset con datos históricos de terrorismo desde el año 1970. 

El dataset contiene es un archivo csv con distintos atributos para cada atentado registrado.

Enlace dataset: https://www.kaggle.com/START-UMD/gtd/downloads/globalterrorismdb_0718dist.csv/3

### 1.1 Preprocesamiento y guardado de datos necesarios.

<font color='blue'>Primero cargamos los datos del csv. Tenemos en cuenta la codificación para que no salte un error.</font>

```python
Dataset = pd.read_csv("Datos/datos.csv", encoding = "ISO-8859-1")
```

<font color='blue'>Ahora creamos un nuevo dataframe con los datos necesarios y filtramos las filas que no sirvan para nuestro propósito.</font>

```python
# Creación del dataframe.
Dataset_Final = pd.DataFrame()

# Confección de las columnas con los datos deseados.
Dataset_Final["anyo"] = Dataset["iyear"]
Dataset_Final["pais"] = Dataset["country_txt"].astype("str")
Dataset_Final["region"] =  Dataset["region_txt"].astype("str")
Dataset_Final["nombre_grupo"] = Dataset["gname"].astype("str")
Dataset_Final["num_victimas"] = Dataset["nkill"]

# Contamos el número de registros del dataset
print("Hay " + str(len(Dataset_Final)) + " registros.")
```

<font color='blue'>Como todas las operaciones que vamos a realizar requieren saber el número de víctimas, debemos eliminar los registros que no posean una cifra (nulos y NaN).</font>

```python
# Eliminación de filas con 'num_victimas' a NaN
Dataset_Final = Dataset_Final.drop(Dataset_Final[Dataset_Final['num_victimas'] != Dataset_Final['num_victimas']].index)

# Reseteo del índice.
Dataset_Final = Dataset_Final.reset_index(drop=True)

# Contamos el número de registros del dataset después de haberlo limpiado de NaN.
print("Hay " + str(len(Dataset_Final)) + " registros.")
```

<font color='blue'>Transformamos el tipo de la columna 'num_victimas' a int64.</font>

```python
# Transformación de la columna 'num_victimas' a int64.
Dataset_Final["num_victimas"] = Dataset_Final["num_victimas"].astype("int64")

# Ordenamos los registros por año.
Dataset_Final = Dataset_Final.sort_values(by='anyo')
```

```python
Dataset_Final.head(10)
```

<font color='blue'>Guardamos el dataframe en un csv.</font>

```python
Dataset_Final.to_csv("Datos/datos_deseados.csv", encoding="ISO-8859-1")
```

### 1.2 Resolución del ejercicio y discusión.

<font color='blue'>Carga del fichero de datos limpio.</font>

```python
Dataset_Final = pd.read_csv("Datos/datos_deseados.csv", encoding = "ISO-8859-1", index_col=0)
```

<font color='blue'>Dibujamos el diagrama de línea con el número de víctimas por año.</font>

```python
# Almacenamos los años del dateset sin repetir.
lista_anyos = Dataset_Final["anyo"].unique()

# Agrupamos el dataset por año, sumamos el nº de víctimas y guardamos el resultado.
lista_sum_victimas_anyo = np.array(Dataset_Final.groupby("anyo").sum()["num_victimas"].tolist())

# Preparamos el diagrama de línea.
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
ax.plot(lista_anyos, lista_sum_victimas_anyo, label='Victimas por año', color='b', marker='p')
plt.xticks(lista_anyos)
plt.xlabel('Año')
plt.ylabel('Víctimas')
plt.xticks(rotation=90)
plt.grid(True)
plt.title("Víctimas por año")
handles, labels = ax.get_legend_handles_labels()
lgd = ax.legend(handles, labels, loc='upper center', bbox_to_anchor=(1,1))
```

<font color='blue'>Según observamos en el diagrama anterior, efectivamente nos encontramos en la época más sangrienta. Desde 2012, vemos una clara subida en el número de víctimas por terrorismo en el mundo.</font>

<font color='blue'>Dibujamos el diagrama de línea con el número de víctimas por año según la región.</font>

```python
# Obtenemos el listado de regiones sin repetir.
lista_regiones = Dataset_Final["region"].unique()
color=iter(cm.rainbow(np.linspace(0,1,len(lista_regiones))))

# Preparamos el diagrama de línea.
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)

# Para cada región, pintamos su diagrama de línea de un color distinto. CUIDADO, no todos las regiones tienen datos
# para todos los años.
for region in lista_regiones:
    datos_region = Dataset_Final[Dataset_Final["region"] == region]
    datos_anyo_region = np.array([])
    for anyo in lista_anyos:
        datos_anyo_region = np.append(datos_anyo_region, datos_region[datos_region["anyo"] == anyo].sum()["num_victimas"])
    ax.plot(lista_anyos, datos_anyo_region, label=region, color=next(color), marker='p')

# Mostramos la gráfica.   
plt.xticks(lista_anyos)
plt.xlabel('Año')
plt.ylabel('Víctimas')
plt.title("Víctimas por año y regiones")
plt.xticks(rotation=90)
plt.grid(True)
handles, labels = ax.get_legend_handles_labels()
lgd = ax.legend(handles, labels, loc='upper center', bbox_to_anchor=(1.05,1), prop={'size': 11})
```

<font color='blue'>Observamos varias épocas clave:
    <br>
    <ul>
        <li>Centroamerica 1980 - 1992</li>
        <li>Norteamérica  2001</li>
        <li>Sur de Asia 2003 - 2017</li>
        <li>Oriente próximo 2003 - 2017</li>
        <li>África subsahariana 2005 - 2017</li>
    </ul>
</font>

#### <font color='blue'>Centroamerica 1980 - 1992 </font>

<font color='blue'>Sacamos los datos de la época para saber mejor qué países sufrieron más el terrorismo.</font>

```python
# Filtramos por la región y ordenamos por el número de victimas.
Centroamerica = Dataset_Final[Dataset_Final["region"] == "Central America & Caribbean"]

# Filtramos por la época y calculamos el número de víctimas por país agrupando por país, sumando el número de víctimas
# y ordenamos de mayor a menor.
Centroamerica = Centroamerica[(Centroamerica["anyo"] <= 1992) & (Centroamerica["anyo"] >= 1980)].groupby("pais").sum() \
    .sort_values(by='num_victimas', ascending=False).head(10)

# Preparamos el diagrama de barras.
y = np.arange(len(Centroamerica.index))
x = Centroamerica["num_victimas"]

# Mostramos la gráfica.
fig, ax = plt.subplots(figsize=(20, 10))
plt.bar(y, x, align='center', alpha=0.5)
plt.xticks(y, Centroamerica.index)
plt.ylabel('Víctimas')
plt.xlabel('País')
plt.title('Terrorismo por países (Centroamerica)')
plt.show()
```

<font color='blue'>Observamos que esta época fue la más sangrienta en Centroamerica debido a las guerras civiles de El Salvador, Nicaragua y Guatemala.</font>

#### <font color='blue'>Norteamérica 2001, Oriente próximo 2003 - 2017</font>

<font color='blue'>Sacamos los datos de la época para saber mejor qué países sufrieron más el terrorismo.</font>

```python
# Filtramos por la región y ordenamos por el número de victimas.
OrienteProximo = Dataset_Final[Dataset_Final["region"] == "Middle East & North Africa"]

# Filtramos por la época y calculamos el número de víctimas por país agrupando por país, sumando el número de víctimas
# y ordenamos de mayor a menor.
OrienteProximo = OrienteProximo[(OrienteProximo["anyo"] >= 2003) & (OrienteProximo["anyo"] <= 2017)].groupby("pais").sum() \
    .sort_values(by='num_victimas', ascending=False).head(10)

# Preparamos el diagrama de barras.
y = np.arange(len(OrienteProximo.index))
x = OrienteProximo["num_victimas"]

# Mostramos la gráfica.
fig, ax = plt.subplots(figsize=(20, 10))
plt.bar(y, x, align='center', alpha=0.5)
plt.xticks(y, OrienteProximo.index)
plt.ylabel('Víctimas')
plt.xlabel('País')
plt.title('Terrorismo por países (Norteamérica)')
plt.show()
```

<font color='blue'>En Norteamérica en 2001 tuvo lugar el atentado conocido como 11S. Este evento supuso la invasión de Iraq (Oriente próximo) por parte de EEUU y algunos países de Europa, lo que supuso un aumento de la tensión en la zona que propició un auge de terrorismo contra las fuerzas occidentales y contra sus simpatizantes. El vacío de poder provocó el surgimiento de ISIS que, junto a la guerra civil en Siria, hizo que los números se acrecentaran.</font>

## Ejercicio 2

#### Otro dato interesante sería conocer cual ha sido el grupo terrorista que más victimas ha causado. Para ello obtén un pie chart con la información de los 10 grupos más sanguinarios de la historia. Interpreta los datos.

```python
color=iter(cm.rainbow(np.linspace(0,1,10)))

# Obviamos los atentados de los que no se sepa el grupo terrorista responsable, agrupamos por grupo, sumamos sus
# víctimas y ordenamos de mayor a menor.
top_10_victimas = Dataset_Final[Dataset_Final["nombre_grupo"] != "Unknown"].groupby("nombre_grupo").sum() \
    .sort_values(by='num_victimas', ascending=False).head(10)

# Preparamos la gráfica.
etiquetas = top_10_victimas.index
tamanyos = top_10_victimas["num_victimas"]

# Mostramos la gráfica.
fig, ax = plt.subplots(figsize=(20, 10))
ax.pie(tamanyos, labels=etiquetas, autopct='%1.1f%%', shadow=True, startangle=90, colors=color)
ax.axis('equal')  
plt.title("Nº de víctimas por grupo")
plt.show()
```

<font color='blue'>El ISIS (Iraq), los talibanes (Afganistan) y Boko Haram (Nigeria) son los grupos más sanguinarios y lideran la cifra de víctimas en sus zonas.</font>

#### Siempre es interesante cruzar los datos de un dataset con otro. A priori parece ser que muchas veces el terrorismo surge con más fuerza en épocas de crisis. Podemos dibujar en un diagrama de dispersión el número de victimas con el PIB de un país. Para ello encuentra un dataset que tenga los PIBs de los distintos países. Dibuja dos diagramas de dispersión: uno con los 10 países con el PIB más bajo en el último año y el número de victimas en atentados y otro con los 10 países con más victimas en el último año y sus PIBs. Interpreta los datos.

#### <font color='blue'>Estudio de 10 países con el PIB más bajo en el último año y el número de victimas en atentados.</font>

<font color='blue'>Carga, limpieza y guardado del dataset de PIB. FUENTE: Banco Mundial.</font>

```python
# Carga del dataset.
Dataset_PIB = pd.read_csv("Datos/GDP.csv", header=3, encoding="ISO-8859-1")

# Confección de las columnas con los datos deseados.
Dataset_PIB_Limpio = pd.DataFrame()
Dataset_PIB_Limpio["pais"] = Dataset_PIB["Economy"]
Dataset_PIB_Limpio["pib"] = Dataset_PIB["US dollars)"].fillna(-1)
Dataset_PIB_Limpio = Dataset_PIB_Limpio.drop(Dataset_PIB_Limpio[Dataset_PIB_Limpio['pib'] == -1].index)
Dataset_PIB_Limpio = Dataset_PIB_Limpio.drop(Dataset_PIB_Limpio[Dataset_PIB_Limpio['pib'] == " .. "].index)
Dataset_PIB_Limpio["pib"] = Dataset_PIB_Limpio["pib"].str.replace(',', '').astype("int64")
Dataset_PIB_Limpio = Dataset_PIB_Limpio.sort_values(by=["pib"], ascending=False)
Dataset_PIB_Limpio = Dataset_PIB_Limpio.reset_index(drop=True)

# Guardamos el dataframe en un csv. 
Dataset_PIB_Limpio.to_csv("Datos/pib_limpio.csv", encoding="ISO-8859-1")

# Mostramos los primeros 10 registros.
Dataset_PIB_Limpio.head(10)
```

<font color='blue'>Filtramos por el último año.</font>

```python
Dataset_Ultimo_anyo = Dataset_Final[Dataset_Final["anyo"] == 2017]
```

<font color='blue'>Cargamos los datos del PIB en el último año.</font>

```python
Dataset_PIB_Limpio = pd.read_csv("Datos/pib_limpio.csv", encoding="ISO-8859-1", index_col=0)
```

<font color='blue'>Muestra de los datos. Ampliamos la muestra a los 75 países con menos PIB para poder tener más referencias.</font>

```python
# Ordenamos por PIB de menor a mayor.
Dataset_PIB_Limpio = Dataset_PIB_Limpio.sort_values(by='pib', ascending=True)

pib_pais = []
victimas_pais = []
paises = []

i = 0

# Tenemos que buscar los países que estén en ambos datasets (enlazamos por el nombre del país en inglés).
for pais in Dataset_PIB_Limpio['pais']:
    if i == 75:
        break
    Dataset_Temp = Dataset_Ultimo_anyo[Dataset_Ultimo_anyo["pais"] == pais]
    
    # Si está en el dataset, lo añadimos a la lista.
    if (len(Dataset_Temp) > 0):
        victimas_pais_temp = Dataset_Ultimo_anyo[Dataset_Ultimo_anyo["pais"] == pais].sum()["num_victimas"]
        victimas_pais = np.append(victimas_pais, victimas_pais_temp)
        pib_pais = np.append(pib_pais, Dataset_PIB_Limpio[Dataset_PIB_Limpio["pais"] == pais]["pib"])
        paises = np.append(paises, pais)
        i += 1 
        
# Preparamos la gráfica.
x = pib_pais
y = victimas_pais

fig, ax = plt.subplots(figsize=(20, 10))
plt.scatter(x,y,c="b")

# Pintamos los nombre de los países.
for i, txt in enumerate(paises):
    pos = (x[i] + 5000, y[i])
    ax.annotate(txt, (x[i], y[i]), pos, 'data', \
                size=10)

# Mostramos la gráfica.
plt.xlabel('PIB')
plt.ylabel('Víctimas')
plt.title('10 países con el PIB más bajo en el último año y el número de victimas')
plt.grid(True)
plt.show()
```

<font color='blue'>Parace que los países con menos economía sufren más los azotes del terrorismo. Observamos también que los países con más víctimas se encuentran también entre los países más pobres.</font>

#### <font color='blue'>Estudio de 10 países con más victimas en el último año y sus PIBs</font>

```python
Paises_Mas_Victimas = pd.DataFrame()

Paises_Mas_Victimas["pais"] = Dataset_Final[Dataset_Final["anyo"] == 2017].groupby("pais").sum() \
    .sort_values(by='num_victimas', ascending=False).head(100).index

Paises_Mas_Victimas["num_victimas"] = Dataset_Final[Dataset_Final["anyo"] == 2017].groupby("pais").sum() \
    .sort_values(by='num_victimas', ascending=False).head(100)["num_victimas"].tolist()

Paises_Mas_Victimas.head(10)
```

```python
pib_pais = []
victimas_pais = []
paises = []

j = 0

# Tenemos que buscar los países que estén en ambos datasets (enlazamos por el nombre del país en inglés).
for pais in Paises_Mas_Victimas['pais']:
    if j == 10:
        break
    Dataset_Temp = Paises_Mas_Victimas[Paises_Mas_Victimas["pais"] == pais]
    
    # Si está en el dataset, lo añadimos a la lista.
    if (len(Dataset_Temp) > 0 and len(Dataset_PIB_Limpio[Dataset_PIB_Limpio["pais"] == pais]["pib"]) >0):
        victimas_pais = np.append(victimas_pais, Paises_Mas_Victimas[Paises_Mas_Victimas['pais'] == pais]["num_victimas"])
        pib_pais = np.append(pib_pais, Dataset_PIB_Limpio[Dataset_PIB_Limpio["pais"] == pais]["pib"])
        paises = np.append(paises, pais)
        j += 1 

# Preparamos la gráfica. 
x = pib_pais
y = victimas_pais

fig, ax = plt.subplots(figsize=(20, 10))
plt.scatter(x,y,c="b")

# Pintamos los nombre de los países.
for i, txt in enumerate(paises):
    pos = (x[i] + 25, y[i])
    ax.annotate(txt, (x[i], y[i]), pos, 'data', \
                size=10)

# Mostramos la gráfica.
plt.xlabel('PIB')
plt.ylabel('Víctimas')
plt.title('10 países con más victimas en el último año y sus PIBs')
plt.grid(True)
plt.show()
```

<font color='blue'>Confirmamos que países como Iraq, Afganistán, Somalia o Nigeria son los países que en el último año tuvieron más víctimas y además están entre los países con menos PIB.</font>
