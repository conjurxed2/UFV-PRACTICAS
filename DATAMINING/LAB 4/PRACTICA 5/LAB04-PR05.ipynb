#### Librerias

```python
import pandas as pd
import matplotlib.pyplot as plt  
import numpy as np  
import warnings
import numpy as np
from pylab import *
from sklearn.cluster import KMeans  
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn import metrics
from sklearn.cluster import AgglomerativeClustering
from scipy.spatial.distance import cdist
warnings.filterwarnings('ignore')
```

#### Funciones

```python
DIR_DATASET = "Datos/zoo.csv"
```

## 1. Preparación del dataset

### 1.1 Carga y limpieza de campos nulos

```python
Datos = pd.read_csv(DIR_DATASET)
print("Hay", len(Datos), "registros.")
nulos = pd.DataFrame(Datos.isnull().sum(), columns=["Número de nulos"])
nulos[nulos["Número de nulos"] != 0]
```

<font color="blue">No hay campos nulos.</font>

```python
Datos.head()
```

### 1.2 Preparación de los datos
<br>
<font color="blue">Centramos los datos y reducimos la dimensionalidad para poder pintar un mapa en 2 dimensiones.</font>

```python
# Clonamos el dataset.
datos_temp = Datos[:]

# Eliminamos la columna de nombre.
del datos_temp["animal_name"]

# Centramos los datos.
datos_centrados = StandardScaler().fit_transform(datos_temp.values)

# Usamos la libreria de sklearn par proyectar los datos.
modelo_PCA = PCA(n_components=2)

# Capturamos las componentes principales.
componentes_principales = modelo_PCA.fit_transform(datos_centrados)

# Creamos el dataframe.
AutoValores = pd.DataFrame()

# Confeccionamos el dataframe con las columnas deseadas.
AutoValores["autovalor"] = modelo_PCA.explained_variance_
AutoValores["% varianza"] = modelo_PCA.explained_variance_ratio_ * 100
AutoValores["% acum"] = modelo_PCA.explained_variance_ratio_.cumsum() * 100

# Mostramos el dataframe.
AutoValores
```

<font color="blue">Con 2 componentes podemos representar el conjunto de datos con un 50% de representatividad.</font>

### 1.3 Cálculo de la proyección en 2D

```python
# Creamos el dataframe con las componentes principales calculadas para cada registro.
Df_Componentes_Principales = pd.DataFrame(data=componentes_principales, columns=["comp_1", "comp_2"])

# Añadimos la columna con el nombre del animal.
Df_Componentes_Principales["animal"] = Datos["animal_name"]

# Mostramos el dataframe.
Df_Componentes_Principales.head()
```

## 2. Construcción del modelo

<font color="blue">Construcción de un modelo para calcular la mejor K.</font>

```python
# Distancias de cada modelo.
distancias = {}

# Bucle de creación de modelos con K distintos.
for K in range(1, 10):
    
    print("Probando modelo con K=", K)
    print("--------------------------")
    
    # Creación del modelo.
    modelo = KMeans(n_clusters=K, max_iter=50)
    
    # Entrenamiento del modelo.
    modelo.fit(componentes_principales)
    
    # Predicción de los grupos del dataset.
    prediccion = modelo.predict(componentes_principales)
    
    # Dibujamos el mapa de agrupaciones.
    fig, ax = plt.subplots(figsize=(20, 10))
    plt.scatter(Df_Componentes_Principales["comp_1"], Df_Componentes_Principales["comp_2"], \
                c=prediccion, s=50, cmap='viridis')
    for i, txt in enumerate(Df_Componentes_Principales["animal"]):
        pos = (Df_Componentes_Principales["comp_1"][i] + 0.05, Df_Componentes_Principales["comp_2"][i])
        ax.annotate(txt, (Df_Componentes_Principales["comp_1"][i], Df_Componentes_Principales["comp_2"][i]), \
                    pos, 'data', size=12)

    # Calculamos los centroides y los añadimos al mapa.
    centroides = modelo.cluster_centers_
    plt.scatter(centroides[:, 0], centroides[:, 1], c='red', s=200, alpha=0.5)
    plt.grid(True)
    plt.xlabel('comp_1')
    plt.ylabel('comp_2')
    plt.title('Datos proyectados en 2 dimensiones')
    plt.show()
    
    # Guardamos la suma del cuadrado de las distancias de las muestras al centroide más cercano.
    distancias[K] = modelo.inertia_
```

<font color="blue">Creamos una gráfica con la evolución de la suma del cuadrado de las distancias de las muestras al centroide más cercano, para decidir el parámetro K óptimo mediante el método del "codo". En esté método, K representa el punto donde se comienzan a separar los distintos grupos en exceso.</font>

```python
sfig, ax = plt.subplots(figsize=(20, 10))
plt.plot(list(distancias.keys()), list(distancias.values()), 'bo-')
plt.grid(True)
plt.title("Evolución de la suma del cuadrado de las distancias de las muestras al centroide más cercano")
plt.xlabel("Número de grupos")
plt.ylabel("Suma de distancias")
plt.show()
```

<font color="blue">Observando la gráfica, decidimos tomar con parámetro K el valor 7. A partir de K=7 los grupos comienzas a separarse en exceso y con esa K encontramos grupos de animales, que interpretados, tienen mucho sentido.</font>

```python
# Creación del modelo.
modelo = KMeans(n_clusters=7, max_iter=50)

# Entrenamiento del modelo.
modelo.fit(componentes_principales)

# Predicción de los grupos del dataset.
predecido = modelo.predict(componentes_principales)

# Dibujamos el mapa de agrupaciones.
fig, ax = plt.subplots(figsize=(20, 10))
plt.scatter(Df_Componentes_Principales["comp_1"], Df_Componentes_Principales["comp_2"], \
            c=predecido, s=50, cmap='viridis')
for i, txt in enumerate(Df_Componentes_Principales["animal"]):
    pos = (Df_Componentes_Principales["comp_1"][i] + 0.05, Df_Componentes_Principales["comp_2"][i])
    ax.annotate(txt, (Df_Componentes_Principales["comp_1"][i], Df_Componentes_Principales["comp_2"][i]), \
                pos, 'data', size=12)

# Calculamos los centroides y los añadimos al mapa.
centroides = modelo.cluster_centers_
plt.scatter(centroides[:, 0], centroides[:, 1], c='red', s=200, alpha=0.5)
plt.grid(True)
plt.xlabel('comp_1')
plt.ylabel('comp_2')
plt.title('Datos proyectados en 2 dimensiones')
plt.show()
```

<font color="blue">Observamos:
    <br>
<ul>
    <li>Verde azulado: Grupo compacto y diferenciado de mamíferos acuáticos como la foca o el león marino.</li>
    <li>Morado oscuro: Grupo compacto y diferenciado de mamíferos terrestres. En estre grupo se ha incluido el ornitorrinco, pero como vemos, está a medio camino del grupo de aves. Recordemos que el ornotorrinco es un mamífero que pone huevos, por lo que tiene sentido que esté solo y entre grupos.</li>
    <li>Amarillo: Grupo compacto donde están los insectos.</li>
    <li>Verde cclaro: Grupo compacto aunque con más error que el resto. Aquí se encuentran los anfibios.</li>
    <li>Azul oscuro: Grupo compacto donde están los peces.</li>
    <li>Morado claro: Grupo menos compacto. Están las aves y están algunos animales que, en principio, no encajan en el grupo como la tortuga, la ostra o el escorpión. El error en estre grupo es alto.</li>
    <li>Azul verdoso: Grupo compacto donde están los crustáceos. En ese grupo, la medusa se encuentra separado del resto, pues sus características son más diferentes.</li>

    <br>
Nos damos cuenta de que los mamíferos se encuentran a la izquierda del eje Y y el resto a la derecha. También, los animales más acuáticos (como los mamíferos acuáticos y los peces) se encuentran en cotas del eje Y más altas que los terrestres (como los insectos).

## 3. Dendograma con clustering jerárquico

<font color="blue">Dibujamos una línea que indica el nivel donde se encuentran los grupos deseados (7).</font>

```python
def GetEtiqueta(Index):
    """
    Función para obtener una etiqueta dado el índice de la misma.
    :param Index: Índice de la etiqueta.
    :return: Devuelve la etiqueta. 
    """
    return etiquetas[Index]

linked = linkage(componentes_principales, 'ward')

# Etiquetas.
etiquetas = Df_Componentes_Principales["animal"]

# Dibujo del dendograma.
plt.figure(figsize=(20,10))
plt.title('Dendograma', fontsize=20)
plt.xlabel('Etiquetas', fontsize=16)
plt.ylabel('Distancia', fontsize=16)

dendrogram(linked, leaf_label_func=GetEtiqueta, leaf_rotation=90., leaf_font_size=10.)

# Dibujamos una línea horizontal que para indicar el nivel de detalle deseado (6 grupos).
plt.axhline(y=4)
plt.show()
```

<font color="blue">Observamos agrupaciones interesantes:
<ul>
    <li>Existe un grupo formado por los mamíferos acuáticos, como el delfín, foca o la marsopa. Este grupo se une con el grupo más genérico de mamíferos.</li>
    <li>Para el caso del ornitorrinco, este se sitúa entre los mamíferos y los animales que ponen huevos.</li>
    <li>Los peces tienen un grupo muy definido, al igual que los insectos.</li>
</ul>
