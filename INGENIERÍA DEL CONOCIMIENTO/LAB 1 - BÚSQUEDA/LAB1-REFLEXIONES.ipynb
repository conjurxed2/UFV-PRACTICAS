### <font color="blue">Consideraciones sobre el ejercicio:</font>
#### <font color="blue">Costes y distancias:</font>

<font color="blue">
    <ul>
        <li>
            La distancia horizontal/vertical es 1 \* coste de la celda en la que estamos.
        </li>
        <li>
            La distancia en diagonal es sqrt(2) \* coste de la celda en la que estamos.
        </li>
        <li>
            El coste de todas las celdas transitables es 1.
        </li>
    </ul>
</font>

#### <font color="blue">Celdas transitables:</font>

<font color="blue">
    <ul>
        <li>
            Árboles: no transitables.
        </li>
        <li>
           Pantano: transitable y accesible desde terreno.
        </li>
        <li>
            Agua: transitable si estás dentro, pero no accesible desde terreno.
        </li>
    </ul>
</font>

#### <font color="blue">Suposiciones:</font>

<font color="blue">
    <ul>
        <li>
            Pantano: accesible desde terreno y desde el agua.
        </li>
        <li>
            Terreno: accesible desde pantano
        </li>
        <li>
            Agua: accesible desde solo el pantano
        </li>
    </ul>
    De esta forma aclaramos que para ir del agua al terreno, o viceversa, hay que pasar por el pantano.
</font>
