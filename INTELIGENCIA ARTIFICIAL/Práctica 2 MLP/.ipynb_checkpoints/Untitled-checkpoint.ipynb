```python
# Aqu√≠ se importan las librerias que se van a utiliar

%reset

import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import plot_model
import keras.optimizers


%matplotlib inline
```

```python
entradas_XOR = np.array([[0,0],[0,1],[1,0],[1,1]])
resultados_XOR = np.array([[0],[1],[1],[0]])

model = Sequential()
model.add(Dense(units=16, activation='relu', input_dim=2))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(lr = 0.1))

model.fit(entradas_XOR, resultados_XOR, epochs=500, verbose=0)

print(model.evaluate(entradas_XOR, resultados_XOR))
print(model.predict(entradas_XOR))

plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)


```

## Parte 2

```python
entrada = np.array(pd.read_csv('Datos/p2_entrenamiento.csv', header=None))
salida_esperada = np.array(pd.read_csv('Datos/p2_objetivos.csv', header=None))

model = Sequential()
model.add(Dense(units=30, activation='relu', input_dim=9))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(lr = 0.01))

model.fit(entrada, salida_esperada, epochs=500, verbose=0)

print(model.evaluate(entrada, salida_esperada))
print(model.predict(entrada))
print(model.get_weights())
```
