```python
%reset
#Imports Api keras
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras import losses
from keras.optimizers import Adam

#Imports numpy y pandas
import numpy as np
import pandas as pd


#Datasets con nuestros datos de los ficheros csv
datos = pd.read_csv('../Datos/p2_entrenamiento.csv', header=None)
datos2 = pd.read_csv('../Datos/p2_objetivos.csv',header=None)

x=[]
y=[]
for i in range(len(datos)):
    x.append((datos.loc[i,0],datos.loc[i,1],datos.loc[i,2],datos.loc[i,3],datos.loc[i,4],
             datos.loc[i,5],datos.loc[i,6],datos.loc[i,7],datos.loc[i,8]))


for i in range(len(datos2)):
    y.append((datos.loc[i,0]))

#Datasets en Arrays numpy
x_train_p2 = np.array(x)
y_train_p2 = np.array(y)


#print(x_train_p2)
#print(y_train_p2)

#Inicializamos el modelo secuencial
model_1_p2 = Sequential()


""" -units: Número de neuronas en la capa oculta
    -activation: Función de activación 
    -input_dim: Número de entradas de la red"""

#Añadimos Capas a nuestro modelo (2 capas)
model_1_p2.add(Dense(300, activation='relu', input_dim=9)) 
model_1_p2.add(Dense(1, activation='sigmoid'))


    
#Optimizador Adam con lr = 0,0000001
model_1_p2.compile(loss='mean_squared_error', optimizer=Adam(lr=0.01), metrics=['accuracy'])


#Entranmos el modelo
entrenamiento_model_1_p2 = model_1_p2.fit(x_train_p2, y_train_p2, epochs=500, verbose=0)

error = entrenamiento_model_1_p2.evaluate(entrada, salida_esperada, verbose= 0)

# Impresión del error en %
print("\nError: %.7f%%" % (error*100) + "\n")

```
